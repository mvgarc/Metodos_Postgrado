En el contexto de una empresa minorista dedicada a la venta de electrodomésticos, se evidencia una deficiencia en la toma de decisiones estratégicas debido a la ausencia de indicadores de gestión. Esta situación genera incertidumbre al momento de planificar, organizar y controlar las operaciones, lo que repercute directamente en la eficiencia, productividad y competitividad del negocio. La carencia de herramientas que permitan medir y monitorear el desempeño limita la capacidad de los directivos para identificar áreas críticas de mejora o tomar decisiones basadas en información objetiva y confiable.

La problemática se agrava en entornos empresariales altamente competitivos, donde las decisiones deben ser ágiles, fundamentadas y orientadas al logro de objetivos claros. La falta de indicadores no solo dificulta el control y la evaluación de resultados, sino que también impide una alineación coherente entre las actividades operativas y los lineamientos estratégicos de la organización. Esto puede traducirse en pérdida de oportunidades, baja capacidad de respuesta ante cambios del mercado y desorganización interna.

Es por ello, que se observa un impacto negativo en el rendimiento global de la empresa y en su capacidad para sostenerse y crecer en el tiempo. La incertidumbre provocada por la ausencia de datos relevantes debilita el liderazgo y la toma de decisiones, comprometiendo la eficiencia operativa y el logro de los objetivos institucionales. Esta situación plantea la necesidad de comprender a fondo los efectos de esta carencia como primer paso hacia la mejora del proceso de gestión.

---



**Variable independiente:** Ausencia de indicadores de gestión  
**Variable dependiente:** Toma de decisiones estratégicas  

**Indicadores seleccionados:**

- Frecuencia de reportes
- Uso de indicadores
- Tiempo de toma de decisiones
- Percepción de calidad en las decisiones

**Instrumento simulado:** Entrevistas estructuradas con 10 gerentes.


```python
# Importar librerías
import pandas as pd

# Crear los datos simulados
data = {
    'Nombre': [f'Gerente {i+1}' for i in range(10)],
    'Frecuencia_Reportes': ['mensual', 'semanal', 'mensual', 'nunca', 'semanal',
                            'mensual', 'nunca', 'mensual', 'semanal', 'semanal'],
    'Uso_Indicadores': ['bajo', 'medio', 'medio', 'bajo', 'alto',
                        'medio', 'bajo', 'medio', 'alto', 'medio'],
    'Tiempo_Toma_Decisiones': [5, 2, 4, 8, 1, 3, 7, 4, 2, 3],  # en días
    'Percepcion_Calidad': ['regular', 'buena', 'buena', 'mala', 'excelente',
                           'buena', 'mala', 'regular', 'buena', 'buena']
}
df = pd.DataFrame(data)
df
```

```python
# Estadísticas descriptivas numéricas
df.describe()
```

```python
# Frecuencias categóricas
df['Frecuencia_Reportes'].value_counts()
df['Uso_Indicadores'].value_counts()
df['Percepcion_Calidad'].value_counts()
```

```python
print("Media:", df['Tiempo_Toma_Decisiones'].mean())
print("Mediana:", df['Tiempo_Toma_Decisiones'].median())
print("Moda:", df['Tiempo_Toma_Decisiones'].mode()[0])
print("Desviación estándar:", df['Tiempo_Toma_Decisiones'].std())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Frecuencia_Reportes', data=df)
plt.title("Frecuencia de Reportes de Desempeño")
plt.show()

sns.countplot(x='Uso_Indicadores', data=df, order=['bajo', 'medio', 'alto'])
plt.title("Uso de Indicadores")
plt.show()

sns.boxplot(y='Tiempo_Toma_Decisiones', data=df)
plt.title("Distribución del Tiempo para Tomar Decisiones")
plt.show()
```

```python
sns.boxplot(x='Uso_Indicadores', y='Tiempo_Toma_Decisiones', data=df, order=['bajo', 'medio', 'alto'])
plt.title("Tiempo de Decisión según Uso de Indicadores")
plt.show()
```

A partir del análisis de los datos simulados, se evidencia que la frecuencia de los reportes de desempeño varía entre semanal, mensual y la ausencia total de reportes, siendo la periodicidad mensual la más común entre los gerentes. Este hallazgo sugiere que no existe un sistema estandarizado ni constante de seguimiento del desempeño, lo cual puede dificultar la toma oportuna de decisiones estratégicas.

En cuanto al uso de indicadores de gestión, la mayoría de los participantes reportó un nivel medio o bajo de utilización. Solo una minoría indicó un uso alto, lo que indica que en general no se aprovechan sistemáticamente las métricas como herramientas de apoyo para la toma de decisiones. Esta falta de integración de indicadores puede limitar la capacidad de los directivos para anticipar problemas o medir el impacto de sus acciones.

Respecto al tiempo requerido para tomar decisiones estratégicas, se observa una tendencia a que quienes reportan un bajo uso de indicadores tardan más días en decidir, mientras que aquellos con un uso alto lo hacen en menor tiempo. Esta relación sugiere que el uso efectivo de indicadores contribuye a una toma de decisiones más ágil y posiblemente más fundamentada.

Finalmente, en la percepción sobre la calidad de las decisiones, la mayoría calificó sus decisiones como buenas, aunque se registraron respuestas negativas (mala o regular) entre quienes también reportaron menor uso de reportes o indicadores. Esto refuerza la hipótesis de que la ausencia de herramientas objetivas impacta la calidad y la eficacia del proceso decisional.

En conjunto, estos resultados permiten afirmar que la falta de indicadores de gestión influye negativamente en la eficiencia y efectividad de las decisiones estratégicas dentro de la organización. Implementar un sistema sistemático de medición y seguimiento puede ser una estrategia clave para mejorar el rendimiento global del negocio.
